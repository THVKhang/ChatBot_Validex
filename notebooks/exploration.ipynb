Notebook demo cho flow parser -> retrieval -> generation.

```python
from pathlib import Path
import sys

# Ensure notebook can import project modules when run from notebooks/
PROJECT_ROOT = Path.cwd().resolve().parent
if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

from app.parser import parse_prompt
from app.retriever import retrieve_top_k
from app.generator import generate_outline, generate_draft

prompt = "Viet cho toi mot blog ve police check la gi"
parsed = parse_prompt(prompt)
data_dir = str(PROJECT_ROOT / "data" / "processed")
docs = retrieve_top_k(parsed.topic, data_dir, top_k=3)
outline = generate_outline(parsed)
draft = generate_draft(parsed, docs)

print(parsed)
print("\nOutline:")
for item in outline:
    print("-", item)
print("\nDraft:\n", draft)
```
